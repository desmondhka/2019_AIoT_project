```
basePage = "https://www.baseball-reference.com"
playerStatPage = "/players/gl.fcgi"
```

```
def getPlayerBattingGamelogs(player_id, playerLink):
    colNames = []

    req = requests.get(playerLink).content.decode('utf-8')                            
    soup = BeautifulSoup(req,'html.parser')
    battingGamelogs = soup.find('table', id="batting_gamelogs")
    #print(battingGamelogs.prettify())

    thead = battingGamelogs.find('thead').find_all('th')
    for th in thead:
        colNames.append(th.string)
    #print(colNames)

    df = pd.DataFrame(columns=colNames)
    cnt = 1

    tbody = battingGamelogs.find('tbody').find_all('tr')
    for tr in tbody:
        rowList = []
        if tr.has_attr("class"):
            if tr['class'][0] == 'thead':
                continue
            elif tr['class'][0] == 'partial_table':
                continue

        for cell in tr.contents:
            rowList.append(cell.string)
            #print(cell.string, end=" ")
        #print("")
        df.loc[cnt] = rowList
        cnt += 1

    df.to_csv("/content/drive/My Drive/gamelogs/2018_b_{}.csv".format(player_id))
```

```
def getPlayerPitchingGamelogs(player_id, playerLink):
    colNames = []

    req = requests.get(playerLink).content.decode('utf-8')                            
    soup = BeautifulSoup(req,'html.parser')
    pitchingGamelogs = soup.find('table', id="pitching_gamelogs")
    #print(pitchingGamelogs.prettify())

    thead = pitchingGamelogs.find('thead').find_all('th')
    for th in thead:
        colNames.append(th.string)
    #print(colNames)

    df = pd.DataFrame(columns=colNames)
    cnt = 1

    tbody = pitchingGamelogs.find('tbody').find_all('tr')
    for tr in tbody:
        rowList = []
        if tr.has_attr("class"):
            if tr['class'][0] == 'thead':
                continue
            elif tr['class'][0] == 'partial_table':
                continue

        for cell in tr.contents:
            rowList.append(cell.string)
            #print(cell.string, end=" ")
        #print("")
        df.loc[cnt] = rowList
        cnt += 1

    df.to_csv("/content/drive/My Drive/gamelogs/2018_p_{}.csv".format(player_id))
```

```
def parseTable(playerType, tablelink):
    if playerType == 'batting':
        tableid = "players_value_batting"
        attr = 'b'
        getGamelogsFunc = getPlayerBattingGamelogs
    elif playerType == 'pitching':
        tableid = "players_value_pitching"
        attr = 'p'
        getGamelogsFunc = getPlayerPitchingGamelogs
    else:
        return

    outString = ""
    cnt = 1

    req = requests.get(tablelink).content.decode('utf-8')                            
    soup = BeautifulSoup(req,'html.parser')
    comments = soup.find_all(string=lambda text: isinstance(text, Comment))
    for c in comments:
        outString += c

    with open("temp.txt", "w") as fp:
        fp.write(outString)

    with open("temp.txt", "r") as fp:
        soup = BeautifulSoup(fp)

    btable = soup.find('table', id=tableid).find('tbody')
    players = btable.find_all('tr', 'full_table')
    totalPlayers = len(players)
    for player in players:
        td = player.find('td')
        playerID = td['data-append-csv']

        getAttr = "?id={}&t={}&year=2018".format(playerID, attr)
        playerLink = basePage + playerStatPage + getAttr
        #print(playerLink)
        getGamelogsFunc(playerID, playerLink)
        
        print("{}\t......\t{}/{}".format(playerID, cnt, totalPlayers))
        time.sleep(2)
        cnt += 1
```
